```python
import pandas as pd
```

```python
df = pd.read_csv('covid_vaccine_statewise.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
first_dose = df.groupby('State')['First Dose Administered'].sum()
print("First Dose Administered (State-Wise): \n", first_dose)
```

```python
import matplotlib.pyplot as plt

first_dose.sort_values(ascending=False).plot(kind='bar', figsize=(12,6), title="First Dose Vaccinations by State")
plt.ylabel("Number of Doses")
plt.xlabel("State")
plt.tight_layout()
plt.show()
```
