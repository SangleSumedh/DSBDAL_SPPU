```python
df['age'] = df['age'].fillna(df['age'].mean()) 
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0]) 
df['deck'] = df['deck'].fillna(df['deck'].mode()[0]) 
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0]) 
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'].astype(str))
df['embarked'] = le.fit_transform(df['embarked'].astype(str))
df['class'] = le.fit_transform(df['class'].astype(str))
df['deck'] = le.fit_transform(df['deck'].astype(str))
df['who'] = le.fit_transform(df['who'].astype(str))
df['embark_town'] = le.fit_transform(df['embark_town'].astype(str))
df['alive'] = le.fit_transform(df['alive'].astype(str))
```
